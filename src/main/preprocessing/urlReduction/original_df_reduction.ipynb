```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('C:/Users/01din/Desktop/data APG/data/data_clean/DataCleaned/Data/allDataWithUrls.csv', usecols=['page_url'])
```

```python
arr = df.page_url.values
new_arr = np.array([0]*len(arr), dtype=object)
```

```python
index = 0
for url in arr:
    # Cut off everythin before a ? - indicates a key specific to a user
    if '?' in url:
        url = url.partition("?")[0]
    if '#' in url:
        url = url.partition("#")[0]
    #Reduce 'foutpagina', which means error, to error
    if 'foutpagina' in url:
        url = 'error'
    #Reduce logout pages to logout
    if 'uitgelogd' in url:
        url = 'logout'
    #Reduce login pages to login
    if 'login' in url:
        url = 'login'
    #Reduce search pages to search
    if '#addsearch=' in url:
        url = 'search'
    new_arr[index] = url
    index+=1
```

```python
df_urls = pd.read_csv('C:/Users/01din\Desktop\data APG/references/references_for_shortened_data/references/url_references_reduced.csv')
df_page_urls = pd.DataFrame(new_arr, columns = ['page_url'])
df_page_urls.tail()
```

```python
map_urls = dict(zip(df_urls.urls, df_urls.id))
print(map_urls)
```

```python
df_page_urls.page_url = df_page_urls.page_url.map(map_urls)
print(df_page_urls)
```
