```python
import pandas as pd
import numpy as np
from ast import literal_eval
```

```python
generic = lambda x: literal_eval(x)

conv = {'url_id_path': generic,
        'seconds_spent_path': generic}

df = pd.read_csv('DATA/Data/testGrouped.csv', converters=conv)
```

```python
url_references = pd.read_csv('DATA/references/url_references_reduced.csv')
```

```python
size = url_references.shape[0]
freq_matrix = np.zeros([size+1, size+1], dtype=int)
```

```python
paths = [
        [2, 7, 4, 3, 4, 1],
        [4, 9, 10, 3, 4, 0, 2, 1],
        [0, 4, 3, 2, 1, 5, 8, 9, 10],
        [10, 3, 4, 3, 7, 2],
        [3, 4],
        [6, 4, 3, 7, 6],
        [3, 0, 4, 3, 4, 7, 9],
        [5, 2, 3, 1, 9, 10, 9, 8],
        [3, 2, 6, 9, 7, 6],
        [4, 3, 4, 9, 10, 3, 10, 2, 6, 7, 4, 3]
        ]

matrix = np.zeros([11, 11], dtype=int)
```

```python
i = 1
for path in df['url_id_path']:
    for index, url_id in enumerate(path):
        if index < len(path)-1:
            freq_matrix[url_id][path[index+1]] += 1
    if i % 100000 == 0:
        print(f"{i} rows done!")
    i+=1

print("done!")
        
```

```python
adj_matrix = np.zeros([size+1, size+1], dtype=int)
for index_i, row in enumerate(freq_matrix):
    for index_j, val in enumerate(row):
        if val != 0:
            adj_matrix[index_i][index_j] = 1
```

```python
freq_df = pd.DataFrame(freq_matrix)
freq_df.to_csv('DATA/matrices/frequency_matrix.csv', index=False, header=False)
adj_df = pd.DataFrame(adj_matrix)
adj_df.to_csv('DATA/matrices/adjacency_matrix.csv', index=False, header=False)
```
