```python
import pandas as pd
import numpy as np
```

```python
url_references = pd.read_csv('DATA/url-reference_new.csv')
url_references
```

```python
# Load the data
dataFrameFile = pd.read_csv('DATA/xar')
dataFrameFile #check data looks good
```

```python
# Remove unnecessary columns
dataFrameFile.drop(['user_id', 'dwh_datum_vanaf', 'dwh_datum_tot', 'dwh_datum_aanmaak', 'dwh_datum_partitie', 'start_date', 'dwh_mog',
             'dwh_scn', 'page_name'], axis=1, inplace=True)
dataFrameFile #check data looks good
```

```python
# Add all unique URL's to the reference file

# Take Unique urls in original dataframe
urls_in_dataset = dataFrameFile['page_url'].unique()
# For each unique url
for url in urls_in_dataset:
    # If it is a logout page, do not check
    if 'uitgelogd' not in url:
        # Check if the url is in the Url-refernce file, if not, add it together with id
        if url not in url_references.urls.values:
            url_references = pd.concat([url_references, pd.DataFrame([[url_references.id.iloc[-1]+1, url]], columns=['id', 'urls'], index=[url_references.id.iloc[-1]])])
            
url_references
```

```python
# Save new url file to csv
url_references.to_csv('DATA/url-reference_new.csv', index=False)
```

```python
# Replace all URL's with their id in the dataframe

#Create dictionary where each unique url is a key and its ID is the value
map_urls = dict(zip(url_references.urls, url_references.id))
#Map urls to ids
dataFrameFile.page_url = dataFrameFile.page_url.map(map_urls)

# Rename column to 'page_url_id'
dataFrameFile.rename(columns={'page_url': 'page_url_id'}, inplace=True)
# Convert column to int
dataFrameFile['page_url_id'] = dataFrameFile['page_url_id'].apply(pd.to_numeric).astype('Int64')

dataFrameFile #check data looks good
```

```python
# Replace all True / False values with 1 / 0
dataFrameFile['personalisation'] = dataFrameFile['personalisation'].astype(int)
dataFrameFile['has_been_in_mijn_omgeving'] = dataFrameFile['has_been_in_mijn_omgeving'].astype(int)
dataFrameFile['is_mijn_omgeving_url'] = dataFrameFile['is_mijn_omgeving_url'].astype(int)
dataFrameFile['landed_in_mijnomgeving'] = dataFrameFile['landed_in_mijnomgeving'].astype(int)

dataFrameFile #check data looks good
```

```python
# Save file back to folder as csv
dataFrameFile.to_csv('DATA/file18.csv', index=False)
```
